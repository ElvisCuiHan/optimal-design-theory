```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.array([1, x, x**2])

def f(x):
    return np.array([x, x**2, x**3])

def m(x):
    return np.outer(f(x),f(x))

def sensitivity(x, info):
    return f(x).T.dot(np.linalg.inv(info)).dot(f(x)) - len(info)
    
```

```python
info = (m(1) + m(3) + m(4))+m(2.75)+m(1.11)+m(4)+m(2.9)+m(1.1)+m(4)+m(2.9)#/3
```

```python
z = np.linspace(0,4,1000)
y = []
for i in z:
    y.append(sensitivity(i, info))
```

```python
print(max(y))
print(z[np.argmax(y)])
plt.plot(z, np.array(y))
plt.show()
```
